```python
import os
import pandas as pd

def update_product_status(sku, new_status, file_path='/content/product_inventory.csv'):
    df = pd.read_csv(file_path, dtype={'sku': str})

    if sku in df['sku'].values:
        df.loc[df['sku'] == sku, 'status'] = new_status
        print(f"Updated status for SKU {sku} to '{new_status}'.")
        df.to_csv(file_path, index=False)
    else:
        print(f"SKU [{sku}] not found in the product inventory.")


def add_new_product(product_dict, file_path='/content/product_inventory.csv'):
    new_product = pd.DataFrame([product_dict])
    new_product.to_csv(file_path, mode='a', index=False, header=False)
    print(f"New product with SKU {product_dict['sku']} added to the inventory.")


def save_sales_invoice(sales_invoice, file_path='/content/sales_report.csv'):
    new_invoice = pd.DataFrame([sales_invoice])
    file_exists = os.path.isfile(file_path)
    new_invoice.to_csv(file_path, mode='a', index=False,  header=not file_exists)
    print(f"New invoice {sales_invoice['invoice_no']} added to the Sales report.")
```
